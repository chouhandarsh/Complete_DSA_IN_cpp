```python

# linear search
lis = eval(input("Enter the list you want: "))
target = int(input("Enter the number: "))
def linear_search(lis,target):
    for i in range(len(lis)):
        if lis[i]==target:
            return i

        else:
            return -1
sol = linear_search(lis,target)
if sol!=-1:
    print("The number is in this list")
    
else:
    print("The number is found in the list at index ")
    

```
